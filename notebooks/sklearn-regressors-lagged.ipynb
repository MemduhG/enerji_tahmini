```python
import pandas as pd

imputed_df = pd.read_csv("imputed.csv").drop("Unnamed: 0", axis=1)
```

```python
lagged = imputed_df.copy()
for col in imputed_df.columns:
    if col in ("year", "hour", "day"):
        continue
    for i in range(1, 7):
        lagged[col + "-" + str(i)] = imputed_df[col].shift(i)

lagged
```

```python
lagged['next_6_hours_production'] = lagged['production'].shift(-1).rolling(window=6, min_periods=1).sum().shift(-6)
#imputed_df['next_6_hours_consumption'] = imputed_df['consumption'].shift(-1).rolling(window=6, min_periods=1).sum().shift(-6)
lagged = lagged[6:-24]
lagged
```

```python

train = lagged[: 365 * 2 * 24]
dev = lagged[365 * 2 * 24: 365 * 2 * 24 + 250 * 24 ]
test = lagged[365 * 2 * 24 + 250 * 24:]

train_x = train.drop("next_6_hours_production", axis=1)
dev_x = dev.drop("next_6_hours_production", axis=1)
test_x = test.drop("next_6_hours_production", axis=1)

train_y = train["next_6_hours_production"]
dev_y = dev["next_6_hours_production"]
test_y = test["next_6_hours_production"]
```
